```python
plt.figure(figsize=(8, 6))
plt.scatter(df['horsepower'], df['acceleration'])
plt.title('Horsepower vs. Acceleration', fontsize=16)
plt.xlabel('Horsepower', fontsize=14)
plt.ylabel('Acceleration', fontsize=14)
plt.grid(True)
plt.show()
```

```python
avg_mpg_by_origin = df.groupby('origin')['mpg'].mean()

plt.figure(figsize=(8, 6))
avg_mpg_by_origin.plot(kind='bar', color=['r', 'g', 'b'], alpha=0.7)
plt.title('Average Fuel Efficiency (mpg) by Country of Origin', fontsize=16)
plt.xlabel('Country of Origin', fontsize=14)
plt.ylabel('Average Fuel Efficiency (mpg)', fontsize=14)
plt.xticks(rotation=0)
plt.grid(True)
plt.show()
```

```python
mpg_usa = df[df['origin'] == 1]['mpg'].tolist()
mpg_germany = df[df['origin'] == 2]['mpg'].tolist()
mpg_japan = df[df['origin'] == 3]['mpg'].tolist()

data = [mpg_usa, mpg_germany, mpg_japan]
labels = ['USA', 'Germany', 'Japan']

plt.figure(figsize=(8, 6))
plt.boxplot(data, labels=labels, patch_artist=True)
plt.title('Fuel Efficiency (mpg) vs. Country of Origin', fontsize=16)
plt.xlabel('Country of Origin', fontsize=14)
plt.ylabel('Fuel Efficiency (mpg)', fontsize=14)
plt.grid(True)
plt.show()
```

```python
mpg_usa = df[df['origin'] == 1]['acceleration'].tolist()
mpg_germany = df[df['origin'] == 2]['acceleration'].tolist()
mpg_japan = df[df['origin'] == 3]['acceleration'].tolist()

data = [mpg_usa, mpg_germany, mpg_japan]
labels = ['USA', 'Germany', 'Japan']

plt.figure(figsize=(8, 6))
plt.boxplot(data, labels=labels, patch_artist=True)
plt.title('Acceleration vs. Country of Origin', fontsize=16)
plt.xlabel('Country of Origin', fontsize=14)
plt.ylabel('Acceleration', fontsize=14)
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
df[df['origin'] == 1]['mpg'].hist(alpha=0.7, label='USA', bins=30)
df[df['origin'] == 2]['mpg'].hist(alpha=0.7, label='Europe', bins=30)
df[df['origin'] == 3]['mpg'].hist(alpha=0.7, label='Asia', bins=30)
plt.title('Distribution of Fuel Efficiency by Region of Origin', fontsize=16)
plt.xlabel('Fuel Efficiency (MPG)', fontsize=14)
plt.ylabel('Frequency', fontsize=14)
plt.legend()
plt.grid(True)
plt.show()
```
